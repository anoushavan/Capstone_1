```python
from datetime import datetime
from collections import defaultdict
from operator import itemgetter
```

```python
# create a defaultdict named date_issued to store the values of the number of permits issued
# on each date
date_issued = defaultdict(int)

# create a series of dates when the permit was approved
dates = dobdata.Approved

# remove all nan values
dates = dates.dropna()

# iterate over the dates series, create a datetime object and add it to the defaultict
for day in dates.values:
    try:
#         day = datetime.strptime(str(day),"%m/%d/%Y")
        date_issued[day] += 1
    except: 'ValueError'        

sorted(date_issued.items(), key=itemgetter(1),reverse = True)
```
